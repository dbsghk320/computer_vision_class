```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('./images/morph_dot.png')

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
erosion = cv2.erode(img, kernel)

merged = np.hstack((img, erosion))


plt.imshow(merged)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img1 = cv2.imread('./images/morph_dot.png', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('./images/morph_hole.png', cv2.IMREAD_GRAYSCALE)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))

opening = cv2.morphologyEx(img1, cv2.MORPH_OPEN,kernel)
closing = cv2.morphologyEx(img2, cv2.MORPH_CLOSE,kernel)

merged1 = np.hstack((img1, opening))
merged2 = np.hstack((img2, opening))
merged3 = np.vstack((merged1, merged2))
plt.figure(dpi=150)
plt.imshow(merged3, cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('./images/moon_gray.jpg')
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (9,9))
                  
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT,kernel)
blackhat = cv2.morphologyEx(img, cv2.MORPH_BLACKHAT,kernel)

merged = np.hstack((img,tophat, blackhat))
plt.figure(dpi=150)
plt.imshow(merged)
plt.axis('off')
plt.show()
```

```python
img = cv2.imread('./images/shapes.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, img_thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)

contours, hierachy = cv2.findContours(
    img_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
)

for contour in contours:
    mmt = cv2.moments(contour)
    cx = int(mmt['m10'] / mmt['m00'])
    cy = int(mmt['m01'] / mmt['m00'])
    area = mmt['m00']
    length = cv2.arcLength(contour, True)
    
    cv2.circle(img, (cx, cy), 5, (0, 255, 255), -1)
    cv2.putText(img, str(area), (cx, cy+20), cv2.FONT_HERSHEY_PLAIN, 1, (0, 0, 255))
    cv2.putText(img, str(length), tuple(contour[0][0]), cv2.FONT_HERSHEY_PLAIN, 1, (0, 0, 255))
    

plt.imshow(img, cmap='gray')
plt.show()
```

```python
img = cv2.imread('./images/bad_rect.png')
img2 = img.copy()


img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
contours, hierachy = cv2.findContours(
    thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
)

contour = contours[0]
epsilon = 0.05 * cv2.arcLength(contour, True)
approx = cv2.approxPolyDP(contour, epsilon, True)

cv2.drawContours(img, [contour], -1, (0, 255, 0), 3)
cv2.drawContours(img2, [approx], -1, (0, 255, 0), 3)

fig, (ax1, ax2) = plt.subplots(ncols=2)
ax1.imshow(img)
ax1.axis('off')
ax2.imshow(img2)
ax2.axis('off')
plt.tight_layout()
plt.show()
```

```python
img = cv2.imread('./images/hand.jpg')
img2 = img.copy()


img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)

contours, hierachy = cv2.findContours(
    thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
)

contour = contours[0]
cv2.drawContours(img, [contour], -1, (0,255,0), 1)

hull = cv2.convexHull(contour)
cv2.drawContours(img2, [hull], -1, (0, 255, 0), 3)

hull2 = cv2.convexHull(contour, returnPoints=False)
defects = cv2.convexityDefects(contour, hull2)
for i in range(defects.shape[0]):
    start, end, farthest, distance = defects[i, 0]
    farthest = tuple(contour[farthest][0])
    dist = distance / 255.0
    if dist > 1:
        cv2.circle(img2, farthest, 3, (0,0,255),-1)

fig, (ax1, ax2) = plt.subplots(ncols=2)
ax1.imshow(img)
ax1.axis('off')
ax2.imshow(img2)
ax2.axis('off')
plt.tight_layout()
plt.show()
```

```python
shapes = cv2.imread('./images/shapestomatch.jpg')
target = cv2.imread('./images/4star.jpg')

shapes_gray = cv2.cvtColor(shapes, cv2.COLOR_BGR2GRAY)
target_gray = cv2.cvtColor(target, cv2.COLOR_BGR2GRAY)

_, target_th = cv2.threshold(target_gray, 127,255, cv2.THRESH_BINARY_INV)
_, shapes_th = cv2.threshold(shapes_gray, 127,255, cv2.THRESH_BINARY_INV)

target_contours, _ = cv2.findContours(target_th, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
shapes_contours, _ = cv2.findContours(shapes_th, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

matchs = []
for ctr in shapes_contours:
    match = cv2.matchShapes(target_contours[0], ctr, cv2.CONTOURS_MATCH_I2,0)
    matchs.append((match,ctr))
    cv2.putText(shapes, f'{match:.3f}', tuple(ctr[0][0]), cv2.FONT_HERSHEY_PLAIN, 1, (0, 0, 255), 1)
    
    
matchs.sort(key=lambda x: x[0])
cv2.drawContours(shapes, [matchs[0][1]], -1, (0, 255, 0), 3)

plt.imshow(shapes)
plt.axis('off')
plt.imshow(shapes)
```

```python
img = cv2.imread('./images/sudoku.jpg')
img2 = img.copy()
print(img.shape)
h, w = img.shape[:2]

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(img_gray, 100, 200)

lines = cv2.HoughLines(edges, 1, np.pi/180, 130)
for line in lines:
    r, theta = line[0]
    tx, ty = np.cos(theta), np.sin(theta)
    x0, y0 = tx*r, ty*r
    cv2.circle(img2, (abs(int(x0)), abs(int(y0))), 3, (0, 0, 255), -1)
    x1, y1 = int(x0+w*(-ty)),int(y0+h*tx)
    x2, y2 = int(x0+w*(+ty)),int(y0-h*tx)
    cv2.line(img2, (x1, y1), (x2,y2), (0,255,0),1)
    
plt.imshow(img2)
plt.show()
```

```python
img = cv2.imread('./images/5shapes.jpg')
img2 = img.copy()
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, th = cv2.threshold(img_gray, 127,255, cv2.THRESH_BINARY_INV)

contours, _ = cv2.findContours(th, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
for contour in contours:
    approx = cv2.approxPolyDP(contour, 0.01*cv2.arcLength(contour, True), True)
    vertices = len(approx)
    print(f'Vertices: {vertices}')
    
    mmt = cv2.moments(contour)
    cx, cy = int(mmt['m10'] / mmt['m00']), int(mmt['m01'] / mmt['m00'])
    
    name = ''
    if vertices == 3:
        name = 'Triangle'
        color = (0, 255, 0)
    elif vertices == 4:
        x, y, w, h = cv2.boundingRect(contour)
        if abs(w-h) <= 3:
            name = 'Square'
            color = (0, 125, 255)
        else:
            name = 'Rectangle'
            color = (0, 0, 255)
    elif vertices == 10:
        name = 'Star'
        color = (255, 255, 0)
    elif vertices == 15:
        name = 'Circle'
        color = (0, 120, 120)
    cv2.drawContours(img2, [contour], -1, color, -1)
    cv2.putText(img2, name, (cx-50, cy), cv2.FONT_HERSHEY_PLAIN, 1, (100,100,100), 1)
    
plt.imshow(img2)
plt.show()
```
