```python
image1 = Image.open('./images/cat-01.jpg')
image2 = Image.open('./images/dog-01.jpg')

fig, axes = plt.subplots(ncols=2)
axes[0].imshow(image1)
axes[1].imshow(image2)
axes[0].set_title(type(image1))
axes[1].set_title(image2.mode)
axes[0].axis('off')
axes[1].axis('off')
fig.tight_layout()
plt.show()
```

```python
import numpy as np

image1 = cv2.imread('./images/cat-01.jpg')
image2 = cv2.imread('./images/dog-01.jpg')

image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

image1 = cv2.resize(image1, (400, 400))
image2 = cv2.resize(image2, (400, 400))

print(image1.shape)
print(image2.shape)

images = np.concatenate((image1, image2), axis=1)
print(images.shape)

plt.imshow(images)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

image = np.full((500, 500, 3), 255, dtype=np.uint8)
cv2.line(image, pt1=(50,50), pt2=(150,50), color=(255, 0, 0), thickness=10)
cv2.rectangle(image, (50, 150), (450, 250), (0, 255, 255), -1)
cv2.putText(image, 'Python', (50, 200), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0))
print(image.shape)
print(image.dtype)
plt.imshow(image)
plt.show()
```
